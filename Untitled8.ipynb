```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
import numpy as np
```

```python
data = fetch_california_housing(as_frame=True)
df = data.frame
df.head()
```

```python
df.describe()
```

```python
X = df.drop(columns=['MedHouseVal'])
y = df['MedHouseVal']
```

1.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['MedInc'], df['MedHouseVal'], alpha=0.4, edgecolor='k')
plt.title('Relação entre Renda Mediana e Valor Mediano das Casas')
plt.xlabel('Renda Mediana (MedInc)')
plt.ylabel('Valor Mediano das Casas (MedHouseVal)')
plt.show()
```

2. 70,30

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

3. Explorando gráficos

```python
plt.figure(figsize=(12, 5))

# Boxplot da renda mediana
plt.subplot(1, 2, 1)
plt.boxplot(df['MedInc'], vert=True, patch_artist=True)
plt.title('Boxplot - Renda Mediana (MedInc)')
plt.ylabel('Renda Mediana')

# Boxplot do valor mediano das casas
plt.subplot(1, 2, 2)
plt.boxplot(df['MedHouseVal'], vert=True, patch_artist=True)
plt.title('Boxplot - Valor Mediano das Casas (MedHouseVal)')
plt.ylabel('Valor Mediano das Casas')

plt.tight_layout()
plt.show()
```

Temos outliers em median income o que faz sentido se vermos a distribuição de renda da população em geral, no valor das casas temos bastante variabilidade dos dados como esperado e uma concentração maior numa faiza de valor mais baixa o que também pode ser relacionado com a distribuição de renda da população, temos poucos outliers mas como a pesquisa n usa uma variável continua de valor e temos um teto, essa representação visual pode não estar fiel a distribuição real dos valores das casas.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
nn_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)
])
```

```python
nn_model.compile(optimizer='adam', loss='mse', metrics=['mae'])


nn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
y_pred_nn = nn_model.predict(X_test).flatten()
```

4.

```python
X_reduced = df[['MedInc', 'HouseAge', 'AveRooms']]
X_train_red, X_test_red, _, _ = train_test_split(X_reduced, y, test_size=0.3, random_state=42)


model_full = LinearRegression()
model_full.fit(X_train, y_train)

model_red = LinearRegression()
model_red.fit(X_train_red, y_train)


y_pred_full = model_full.predict(X_test)
y_pred_red = model_red.predict(X_test_red)

```

```python
mse_full = mean_squared_error(y_test, y_pred_full)
r2_full = r2_score(y_test, y_pred_full)

mse_red = mean_squared_error(y_test, y_pred_red)
r2_red = r2_score(y_test, y_pred_red)
print("Modelo completo:")
print("MSE:", mse_full)
print("R²:", r2_full)

print("\nModelo reduzido:")
print("MSE:", mse_red)
print("R²:", r2_red)
```

```python
plt.figure(figsize=(12, 5))

# Modelo completo
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_full, alpha=0.4, edgecolor='k')
plt.title(f'Modelo Completo (R² = {r2_full:.2f})')
plt.xlabel('Valor Real')
plt.ylabel('Valor Previsto')
plt.plot([0, 5], [0, 5], 'r--')

# Modelo reduzido
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_red, alpha=0.4, edgecolor='k', color='orange')
plt.title(f'Modelo Reduzido (R² = {r2_red:.2f})')
plt.xlabel('Valor Real')
plt.ylabel('Valor Previsto')
plt.plot([0, 5], [0, 5], 'r--')

plt.tight_layout()
plt.show()

# 8️⃣ Comparação visual de desempenho
plt.bar(['Modelo Completo', 'Modelo Reduzido'], [r2_full, r2_red], color=['skyblue', 'orange'])
plt.title('Comparação do Desempenho (R²)')
plt.ylabel('R² Score')
plt.ylim(0, 1)
plt.show()
```

Concluo aqui que o modelo completo ainda representa um pouco melhor a curva por levar em consideração mais featurings para construção da função mas as três colunas usadas na versão reduzida ainda exemplificam bem os dados

Agora vamos para a árvore de decisão e o modelo linear


```python
tree_model = DecisionTreeRegressor(max_depth=10, random_state=42)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
```

```python
def avaliar(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return mse, r2

resultados = {
    "Modelo": ["Regressão Linear", "Árvore de Decisão", "Rede Neural"],
    "MSE": [
        avaliar(y_test, y_pred_full)[0],
        avaliar(y_test, y_pred_tree)[0],
        avaliar(y_test, y_pred_nn)[0]
    ],
    "R²": [
        avaliar(y_test, y_pred_full)[1],
        avaliar(y_test, y_pred_tree)[1],
        avaliar(y_test, y_pred_nn)[1]
    ]
}

df_resultados = pd.DataFrame(resultados)
print("📊 Comparativo de desempenho:")
print(df_resultados)

# === 8. Gráfico de dispersão Real x Previsto ===
plt.figure(figsize=(12,5))
plt.scatter(y_test, y_pred_full, alpha=0.4, label="Regressão Linear")
plt.scatter(y_test, y_pred_tree, alpha=0.4, label="Árvore de Decisão")
plt.scatter(y_test, y_pred_nn, alpha=0.4, label="Rede Neural")
plt.plot([0,5],[0,5],'r--')
plt.xlabel("Valores reais")
plt.ylabel("Valores previstos")
plt.title("Comparação de modelos: Real x Previsto")
plt.legend()
plt.show()

# === 9. Gráfico comparativo de R² e MSE ===
plt.figure(figsize=(10,5))
bar_width = 0.35
x = np.arange(len(df_resultados["Modelo"]))
plt.bar(x - bar_width/2, df_resultados["MSE"], bar_width, label="MSE")
plt.bar(x + bar_width/2, df_resultados["R²"], bar_width, label="R²")
plt.xticks(x, df_resultados["Modelo"])
plt.ylabel("Valor")
plt.title("Comparação de desempenho dos modelos")
plt.legend()
plt.show()
```

Os resultados da comparação entre os três modelos mostram diferenças claras em termos de desempenho. O modelo de Árvore de Decisão apresentou o menor erro médio quadrático (MSE = 0.416) e o maior coeficiente de determinação (R² = 0.683), indicando que é capaz de explicar aproximadamente 68% da variabilidade dos valores das casas. Isso significa que ele ajusta melhor os dados, conseguindo capturar relações não lineares e interações entre variáveis.
A Rede Neural teve desempenho intermediário, com MSE = 0.507 e R² = 0.614, explicando cerca de 61% da variação dos preços das casas. Embora seja capaz de modelar padrões não lineares, seu desempenho depende de ajustes como o número de camadas, neurônios e épocas de treinamento, além da padronização dos dados.
A Regressão Linear apresentou o pior desempenho entre os três, com MSE = 0.531 e R² = 0.596, explicando cerca de 60% da variabilidade dos preços. Isso ocorre porque ela assume uma relação linear entre as variáveis independentes e o valor das casas, não conseguindo capturar interações complexas ou padrões não lineares presentes no dataset.
Em resumo, a Árvore de Decisão foi a que melhor se ajustou aos dados, seguida pela Rede Neural. A Regressão Linear, apesar de menos precisa, continua útil como modelo baseline devido à sua simplicidade e interpretabilidade. Em aplicações práticas, muitas vezes combina-se modelos lineares para explicabilidade com modelos mais complexos, como árvores ou redes neurais, para melhorar a precisão preditiva.
